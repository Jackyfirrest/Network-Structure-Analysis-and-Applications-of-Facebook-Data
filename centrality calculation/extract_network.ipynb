# Convert to matrix indices

The code below converts the networks to matrix indices. It is unnecessary if using the package `networkx`

```python
user_count = len(df.user_id.unique())
page_count = len(df.page_id.unique())
print(f"Total User:\t{user_count}")
print(f"Total Page:\t{page_count}")

week_time = df.week_start_date.unique()
```

```python
# order the user/page id and give them order number

user_list = sorted(df.user_id.unique())
page_list = sorted(df.page_id.unique())
user_count = len(user_list)
page_count = len(page_list)
    
user_order_id = pd.DataFrame({'u_i':[i for i in range(user_count)],'user_id':user_list})
page_order_id = pd.DataFrame({'p_i':[i for i in range(page_count)],'page_id':page_list})
user_order_id.to_csv('DATA/user_list.csv', index = False)
page_order_id.to_csv('DATA/page_list.csv', index = False)
```

```python
# assign the ordered number to data, which is uint16
# This reduces the size and maps it to matrix position

df_matrix_i = (df.sort_values(['week_start_date', 'user_id','page_id'])
               .merge(user_order_id, how = 'left', on='user_id')
               .merge(page_order_id, how = 'left', on='page_id')
               )[['u_i','p_i','week_start_date', 'reaction_time']]
df_matrix_i = df_matrix_i.astype({
    'u_i':'uint32',
    'p_i':'uint16',
    'reaction_time':'uint8'
})
```
